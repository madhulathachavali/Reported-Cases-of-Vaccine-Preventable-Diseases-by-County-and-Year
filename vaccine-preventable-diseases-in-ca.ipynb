```python
# Load the CSV file
data = pd.read_csv('/kaggle/input/disease/izb-cases-by-county-year-2022 (1).csv')
data.head()
```

```python

# Filter the data for Los Angeles County
la_data = data[data['County'] == 'Los Angeles']

# Pivot the data to have years as rows and diseases as columns
la_pivot = la_data.pivot(index='Year', columns='Disease', values='Count')

# Plot the trends for each disease over the years
plt.figure(figsize=(15, 8))
for column in la_pivot.columns:
    plt.plot(la_pivot.index, la_pivot[column], label=column)

plt.title('Trend Analysis of Diseases in Los Angeles County (2001-2022)')
plt.xlabel('Year')
plt.ylabel('Number of Cases')
plt.legend(title='Disease')
plt.grid(True)
plt.show()
```

```python
# Filter the data for Los Angeles County
la_data = data[data['County'] == 'Los Angeles']

# Aggregate the total counts for all diseases over all years
la_all_diseases_totals = la_data.groupby('Disease')['Count'].sum().reset_index()

# Plot the bar chart for all diseases
plt.figure(figsize=(16, 8))
plt.bar(la_all_diseases_totals['Disease'], la_all_diseases_totals['Count'], color='skyblue')
plt.title('Total Cases of All Diseases in Los Angeles County (2001-2022)', fontsize=16)
plt.xlabel('Disease', fontsize=14)
plt.ylabel('Total Count', fontsize=14)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y')

plt.show()
```

```python
# Filter the data for Los Angeles County
la_data = data[data['County'] == 'Los Angeles']

# Aggregate the data for plotting
la_all_diseases_yearly = la_data.groupby(['Year', 'Disease'])['Count'].sum().reset_index()

# Get unique list of diseases
unique_diseases = la_all_diseases_yearly['Disease'].unique()

# Determine the number of rows and columns for subplots
n_cols = 5
n_rows = (len(unique_diseases) + n_cols - 1) // n_cols

# Plot the trends for each disease over the years in separate subplots
fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(n_cols*5, n_rows*5))
fig.suptitle('Yearly Trends of All Diseases in Los Angeles County (2001-2022)', fontsize=16)

for ax, disease in zip(axes.flatten(), unique_diseases):
    disease_data = la_all_diseases_yearly[la_all_diseases_yearly['Disease'] == disease]
    ax.plot(disease_data['Year'], disease_data['Count'], marker='o', linestyle='-')
    ax.set_title(disease, fontsize=14)
    ax.set_xlabel('Year', fontsize=12)
    ax.set_ylabel('Number of Cases', fontsize=12)
    ax.grid(True)
    ax.tick_params(axis='x', rotation=45)

# Remove empty subplots
for ax in axes.flatten()[len(unique_diseases):]:
    ax.remove()

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
